## 1. Data Understanding

### 1.1 Importing Libraries

```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import seaborn as sns

sns.set_palette("viridis")
```

### 1.2 Load Dataset

```python
original = pd.read_csv('./data/AmesHousing.csv')
original
```

### Types

```python
original.dtypes
```

### 1.3 Dataset Shape & Feature Types

```python
rows, cols = original.shape
print(f"Dataset shape: {rows} rows × {cols} columns")
print(original.dtypes.value_counts())
```

### 1.4 Understanding the Dataset

```python
# Unique parcel IDs
pd.unique(original.PID)
```

```python
# Drop 'Order', set PID as index
df = original.drop(columns="Order")
df = df.set_index("PID")
df.head()
```

### Statistical summary for all features

```python
df.describe().T.round(2)
```

```python
# Check data type of target variable
print(df["SalePrice"].dtype)

# If classification, count number of unique classes
# (εδώ το κάνουμε για demonstration)
num_classes = df["SalePrice"].nunique()
print("Number of unique SalePrice values:", num_classes)
```

```python
# Count categorical features
categorical_cols = df.select_dtypes(include="object").columns
print("Categorical columns:", len(categorical_cols))

# Show imbalance for each categorical feature
for col in categorical_cols[:5]:  # δείξε τις πρώτες 5 για αρχή
    print(f"\n{col} value counts (%):")
    print(df[col].value_counts(normalize=True).round(2) * 100)
```

### 1.5 Target Variable (SalePrice)

```python
print("\nTarget variable (SalePrice) statistics:")
print(df["SalePrice"].describe().round(2))
```

#### Boxplot

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["SalePrice"])
plt.title("Boxplot of SalePrice")
plt.xlabel("SalePrice")
plt.tight_layout()
plt.show()
```

#### Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(df["SalePrice"], kde=True, bins=40)
plt.title("Distribution of SalePrice")
plt.xlabel("SalePrice")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

### 1.6 Missing Values

```python
missing_vals = df.isnull().sum()
missing_vals = missing_vals[missing_vals > 0].sort_values(ascending=False)
print(missing_vals)
```
